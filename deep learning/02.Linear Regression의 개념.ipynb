## 2.1 linear regression

- linear regression : 어떻게 될것(linear hypothesis)이라고 예측하는 것
- linear hypothesis : H(x) = Wx + b
    - W, b에 수를 대입하고 H(x)를 통해 가설을 판단한다.
- cost : 각 지점의 (H(x)-y)^2를 합하고 총 지점의 개수로 나눠 구한다.
    이 cost값을 가장 작게하는 W, b가 linear regression이다.
